```python
import numpy as np
import matplotlib.pyplot as plt
```

## オイラー法
$x(t + \Delta t) = x(t) + f(x)\Delta t$

$\dot x = 3x + 1, x(0) = 0$を解く (厳密解: $x(t) = \frac{1}{3}(e^{3t} - 1)$)

```python
delta_t = 0.01
t = 0
x = 0

def f(x):
    return 3*x + 1

max_t = 2

t_hist = [t]
x_hist = [x]
while t < max_t:
    x += f(x) * delta_t
    t += 0.01
    x_hist.append(x)
    t_hist.append(t)
```

```python
x_exact = 1/3 * (np.exp(3*np.array(t_hist)) - 1)


plt.plot(t_hist, x_hist, label="Numeric")
plt.plot(t_hist, x_exact, label='Exact')
plt.title("$\dot x = 3x + 1, x(0) = 0$ (Euler's method)")
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.show()
```

## ルンゲクッタ法
$$
a = f(x_i)\Delta t \\
b = f(x_i + \frac{a}{2})\Delta t \\
c = f(x_i + \frac{b}{2})\Delta t \\
d = f(x_i + \frac{c}{2})\Delta t \\
x_{x+1} = x_i + \frac{a + 2b + 2c + d}{6}
$$

```python
delta_t = 0.01
t = 0
x = 0

def f(x):
    return 3*x + 1

max_t = 2

t_hist = [t]
x_hist = [x]
while t < max_t:
    a = f(x) * delta_t
    b = f(x + a / 2) * delta_t
    c = f(x + b / 2) * delta_t
    d = f(x + c) * delta_t
    x = x + (a + 2*b + 2*b + d) /6
    t += 0.01
    x_hist.append(x)
    t_hist.append(t)
```

```python
x_exact = 1/3 * (np.exp(3*np.array(t_hist)) - 1)


plt.plot(t_hist, x_hist, label="Numeric")
plt.plot(t_hist, x_exact, label='Exact')
plt.title("$\dot x = 3x + 1, x(0) = 0$ (Runge-Kutta method)")
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.show()
```
